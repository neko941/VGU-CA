# Relational Operators

```c++
 #include <stdio.h>

// == 
if (5 == 6)
    printf("5 is equal 6\n");
else
    printf("5 is not equal 6\n");
```

```c++
 #include <stdio.h>

// !=
if (5 != 6)
    printf("5 is not equal 6\n");
else
    printf("5 is equal 6\n");
```

```c++
 #include <stdio.h>

// >
if (5 > 6)
    printf("5 is greater than 6\n");
else
    printf("5 is not greater than 6\n");
```

```c++
 #include <stdio.h>

// <
if (5 < 6)
    printf("5 is less than 6\n");
else
    printf("5 is not less than 6\n");
```

```c++
 #include <stdio.h>

// >= 
if (5 >= 6)
    printf("5 is greater than or equal 6\n");
else
    printf("5 is not greater than or equal 6\n");
```

```c++
#include <stdio.h>

// <= 
if (5 <= 6)
    printf("5 is less than or equal 6\n");
else
    printf("5 is not less than or equal 6\n");
```

# Assignment Operators

```c++
#include <stdio.h>

// +=
int a = 5;
int b = 0;
int c = 0;

b += a;
c = c + a;

printf("b = %d\n", b);
printf("c = %d\n", c);
```

```c++
#include <stdio.h>

// -=
int a = 5;
int b = 0;
int c = 0;

b -= a;
c = c - a;

printf("b = %d\n", b);
printf("c = %d\n", c);
```

```c++
#include <stdio.h>

// *=
int a = 5;
int b = 0;
int c = 0;

b *= a;
c = c * a;

printf("b = %d\n", b);
printf("c = %d\n", c);
```

```c++
#include <stdio.h>

// /=
int a = 5;
float b = 1;
float c = 1;

b /= a;
c = c / a;

printf("b = %.1f\n", b);
printf("c = %.1f\n", c);
```

# Conditional Expression Operators

```c++
#include <stdio.h>

// ?:
int a = 5;
int b;  

b = ((a==5) ? 3 : 2 );

printf("b = %d", b); 
```

```c++
#include <stdio.h>

// ?:
int age = 18;

(age >= 18) 
? printf("eligible for voting") 
: printf("not eligible for voting"); 
```

```c++
#include <stdio.h>

void print_array(int arr[], int size)
{
    for (int i = 0; i < size; i++)
    {
        printf("%d\t", arr[i]);
    }
    puts("");
}

int a[] = {1, 2, 3, 4, 5};
int size = sizeof(a)/sizeof(a[0]);

print_array(a, size);
```

```c++
#include <stdio.h>

void print_array(int * arr, int size)
{
    for (int i = 0; i < size; i++)
    {
        printf("%d\t", arr[i]);
    }
    puts("");
}

int a[] = {1, 2, 3, 4, 5};
int size = sizeof(a)/sizeof(a[0]);

print_array(a, size);
```

```c++
#include <stdio.h>
#include <math.h>

float estimate_pi(float sum, int n,int i)
{
    sum += pow(-1, i+1) / (2*i -1);

    if (i < n)
        estimate_pi(sum, n, i+1);
    else
        return 4*sum;
}

printf("pi = %f\n", estimate_pi(0, 5000, 1));
```

```c++
#include <stdio.h>
#include <math.h>

float estimate_pi(int n)
{
    if (n > 0)
        return pow(-1, n+1) / (2*n-1) + estimate_pi(n-1);
    else
        return 0;
}

printf("pi = %f\n", 4*estimate_pi(5000));
```

```c++
#include <stdio.h>

double factorial(int n)
{
    if (n >= 1)
        return n*factorial(n-1);
    else
        return 1;
}

double estimate_e(int n)
{
   if (n >= 1)
        return 1 / factorial(n) + estimate_e(n-1);
    else
        return 1;
}

printf("%f\n", estimate_e(30));
```
