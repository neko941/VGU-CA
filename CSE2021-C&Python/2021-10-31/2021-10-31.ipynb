# printf

```c++
#include <stdio.h>
/*
 * Format: int printf(const char * format, ... );
 * Purpose: display (format) string
 * Library: stdio
 */

// %c: character
printf ("Characters: %c %c \n", 'a', 65);

// %s: string
printf ("%s \n", "A string");

// %d: signed decimal integer
printf ("Decimals: %d \n", 1977);

// %_d: Preceding with blanks (with _ is the number of blanks)
printf ("Preceding with blanks: %10d \n", 1977);

// %0_d: Preceding with zeros (with _ is the number of zeros)
printf ("Preceding with zeros: %010d \n", 1977);

// %x: unsigned hexadecimal integer
printf ("Some different radices: %x \n", 100);

// %o: unsigned octal
printf ("Some different radices: %o \n", 100);

// %f: float
printf ("Floats: %f \n", 3.1416);

// %._f: float but take _ decumal part (with _ is the number of decumal part)
printf ("Floats: %.2f \n", 3.1416);

// %e: scientific (exponential) notation (lowercase)
printf ("Floats: %e \n", 3.1416);

// %E: scientific (exponential) notation (upppercase)
printf ("Floats: %E \n", 3.1416);

// %+f: float but preceed the result with a plus or minus sign (+ or -) 
// By default, only negative numbers are preceded with a - sign
printf ("Floats: %+f \n", 3.1416);
```

# scanf

```c++
#include <stdio.h>
/*
 * Format: int scanf(const char * format, ... );
 * Purpose: reads data from stdin -> stores them (with parameter format) into the locations pointed by the additional arguments
 * Library: stdio
 */

char name [80];
int age;

printf("Enter your name: ");
scanf("%s", name);  

printf("Enter your age: ");
scanf("%d", &age);

printf("\nMr. %s, you are %d years old.\n", name, age);
```

# fgets

```c++
#include <stdio.h>
/*
 * Format: char * fgets(char *variable, int size, FILE *stream)
 * Purpose: reads data from stdin (with the whitespace) -> stores them into the variable
 * Library: stdio
 */

char name [80];

printf("Enter your full name: ");
fgets(name, 80, stdin);  

printf("\nGood Morning Mr. %s\n", name);
```

# Question 1

Write a program that gets an input number from a user and checks if the input is divisible by 4!

```c++
#include <stdio.h>

// declare a variable
int num;

// get an input number from a user
printf("Please enter a number: \n>> ");
scanf("%d", &num);

// check if the num divisible by 4 or not
if (num%4 == 0)
{
    printf("%d is divisible by 4!!!", num);
}
else
{
    printf("%d is not divisible by 4!!!", num);
}
```

# Question 3

Write a program to check whether a given year (entered from a user) is a leap year or not

```c++
#include <stdio.h>

int year;

printf("Please enter a year \n>> ");
scanf("%d", &year);

if ((year%4 == 0 && year%100 != 0) || (year%400 == 0))
{
    printf("%d is a leap year", year);
}
else
{
  printf("%d is not a leap year", year);
}
```
